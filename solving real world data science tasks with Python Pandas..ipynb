### Task 4 : Add a city columns

```python
#lets use apply method:(why we using apply method , need to know more about lambda)
all_data['City'] = all_data['Purchase Address'].apply(lambda x : x.split(',')[1])

all_data.head()
```

```python
#lets use apply method

def get_city(address):
    return address.split(',')[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x : get_city(x))

all_data.head()
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x : get_city(x)+' '+ get_state(x))

all_data.head()
```

```python
#using another format f" so that we can directly use without adding + value
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_data['City'] = all_data['Purchase Address'].apply(lambda x : f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

### plotting the sales per month on graph

```python
plt.figure(figsize =(12,6))
months = range(1,13)
plt.bar(months , sales_by_months['sales'], color = 'green')
plt.plot(months , sales_by_months['sales'], color = 'red')
plt.xticks(months)
plt.ylabel('sales in USD($)in million')
plt.xlabel('months')
plt.show()
```
